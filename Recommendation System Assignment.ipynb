```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv('Recommendation_book.csv', encoding_errors='replace')
data
```

# EDA

```python
data1=data.drop('Unnamed: 0', axis=1)
data1
```

```python
data1=data.rename({'User.ID': 'userid', 'Unnamed: 0': 'index', 'Book.Title': 'booktitle', 'Book.Rating': 'bookrating'},axis=1)
data1
```

```python
data1.info()
```

```python
data1[data1.duplicated()]
```

```python
#Checking for duplicates
data1[data1.duplicated(subset=["userid","booktitle"])]
```

```python
#Dropping duplicates
data2=data1.drop_duplicates(subset=["userid","booktitle"])
data2
```

```python
len(data2.userid.unique()) 
```

```python
len(data2.booktitle.unique()) 
```

```python
table=data2.pivot(index='userid', columns='booktitle', values='bookrating').reset_index(drop=True)
table
```

```python
table.fillna(0, inplace=True)
table
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
distance=1-pairwise_distances(table.values,metric='cosine')
distance
```

```python
#Store the results in a dataframe
newdata = pd.DataFrame(distance)
```

```python
#Set the index and column names to user ids 
newdata.index = data1.userid.unique()
newdata.columns = data1.userid.unique() 
```

```python
newdata.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(distance, 0)
newdata.iloc[0:5, 0:5]
```
